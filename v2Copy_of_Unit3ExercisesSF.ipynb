## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task9**:

Plot the pdfs (techincally the pmfs) of the following distributions: $Bernoulli(p=0.7)$, $Binomial(n=10,p=0.7)$, $Poisson(μ=7)$, $Geometric(p=0.7)$, $HyperGeometric(N=50,k=35,n=15)$, $Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4])$

**Task10**:

Describe what processes "generate" each of the previous discrete distributions models. Also report their support.

Here's a spicier way to phrase the question: describe these distributions' maximum entropy conditions in english.

Hint: Make sure to read any of the: preliZ documentation, relevant wikipedia articles, or wolfram alpha.


**Bernoulli**:
Simulating trials that can only have either a true/false answer(like a coin flip).  Support: [0,1]
**Binomial**: A sequence of bernoulli experiments and then counting how many successes were in a number of those. Support: [0, infinity)
**Poisson**: Modeling the number of times something will occur with a fixed amount of time, assuming we know the mean rate and they are independent from each other. Support: [0, infinity)
**Geometric**: Simulating bernoulli trials all with the same probability of success up until you get a success. Support: [1, infinity)
**Hypergeometric**: A hypergeometric distribution is generated from a population of a fixed size, where items are sorted into two categories that are basically "success" and "failure". It tells you how likely you are to get a specific amount of successes within a fixed sample within the population(sampled without replacement). Support: [0, infinity)
**Categorical**: A categorical distribution runs a random trial with a set number of outcomes with distinct probabilities, and the result is whatever outcome you get. Support: [0, infinity)
