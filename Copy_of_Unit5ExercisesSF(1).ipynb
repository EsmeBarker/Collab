## From Linear Model to Generalized Linear Model

You'll now fit a linear regression on the height and weight for people or the Kalahari !Kung San people. Height will be the predictor and weight will be the predicted.

Here's the raw data.

```python
#downloads the data from my github
howell = pd.read_csv('https://raw.githubusercontent.com/thedarredondo/data-science-fundamentals/main/Data/howell.csv')
```

Make a scatter plot of height versus weight for the howell data set.

**Task5**:

Create and run a stright line linear regression model on the howell data set with height predicting weight. Visualize your model and its posterior.

```python

with pm.Model() as nh_model:
    Î± = pm.Normal("Î±", mu=0, sigma=3)
    Î² = pm.Normal("Î²", mu=0, sigma=10)
    Ïƒ = pm.HalfNormal("Ïƒ", 1)

    y = pm.Normal("y_pred",
                  mu=Î± + Î² * howell['height'],
                  sigma=Ïƒ,
                  observed=howell['weight'])
    idata_nh_h = pm.sample()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 3))

x_range = np.linspace(howell['height'].min(), howell['height'].max(), 100)

posterior = az.extract(idata_nh_h)

alphas = posterior["Î±"].values[:, np.newaxis]
betas = posterior["Î²"].values[:, np.newaxis]

alpha_m = alphas.mean().item()
beta_m = betas.mean().item()

ax.plot(x_range, alpha_m + beta_m * x_range, c="k", label="Mean regression line")

az.plot_hdi(x_range, (alphas + betas * x_range), ax=ax, color="gray", smooth=False, hdi_prob=0.89)

ax.scatter(howell['height'], howell['weight'], s=10, alpha=0.8, color='blue', label='Data points')

ax.set_xlabel('height')
ax.set_ylabel('weight')
ax.legend()

plt.xlim(howell['height'].min() - 1, howell['height'].max() + 1)
plt.ylim(howell['weight'].min() - 1, howell['weight'].max() + 1)
plt.show()
```

**Task6**:

Run a posterior predictive check for model_howell, plot the posterior predictive distribution, and comment on the fit of the model you made in Task5.

```python
pm.sample_posterior_predictive(
    idata_nh_h, model = nh_model, extend_inferencedata=True
    )
```

```python
az.plot_ppc(idata_nh_h, num_pp_samples=200, colors=["C1", "C0", "C1"])
```

The model had the right shape, but it looks a little bit off. Mostly, its peak just seems too high.

**Task7**:

Create and visualize the posterior of a generalized linear model that improves the on the fit of the model you made in Task5.

```python
with pm.Model() as gl_model:
    Î± = pm.Normal("Î±", mu=0, sigma=3)
    Î² = pm.Normal("Î²", mu=0, sigma=10)
    Ïƒ = pm.HalfNormal("Ïƒ", 1)

    mu_log_weight = Î± + Î² * howell['height']
    y = pm.LogNormal("y_pred",
                     mu=mu_log_weight,
                     sigma=Ïƒ,
                     observed=howell['weight'])
    idata_gl_h = pm.sample()
```

```python

fig, ax = plt.subplots(1, 1, figsize=(12, 3))

x_range = np.linspace(howell['height'].min(), howell['height'].max(), 100)

posterior_gl = az.extract(idata_gl_h)

alphas = posterior_gl["Î±"].values[:, np.newaxis]
betas = posterior_gl["Î²"].values[:, np.newaxis]

alpha_m = alphas.mean().item()
beta_m = betas.mean().item()

ax.plot(x_range, np.exp(alpha_m + beta_m * x_range), c="k", label="Mean regression line")

az.plot_hdi(x_range, np.exp(alphas + betas * x_range), ax=ax, color="gray", smooth=False, hdi_prob=0.89)


ax.scatter(howell['height'], howell['weight'], s=10, alpha=0.8, color='blue', label='Data points')

ax.set_xlabel('height')
ax.set_ylabel('weight')
ax.legend()

plt.xlim(howell['height'].min() - 1, howell['height'].max() + 1)
plt.ylim(howell['weight'].min() - 1, howell['weight'].max() + 1)
plt.show()
```

**Task8**:

Run a posterior predictive check, and visualize the posterior predictive distribution, for the model you made in Task7.

Comment on the fit.

```python
pm.sample_posterior_predictive(idata_gl_h, model=gl_model, extend_inferencedata=True)
```

```python
az.plot_ppc(idata_gl_h, num_pp_samples=200, colors=["C1", "C0", "C1"])
plt.show()
```

This fit is better than the last one. The peak is a little low now, but overall it seems like it predicts the data pretty well.
